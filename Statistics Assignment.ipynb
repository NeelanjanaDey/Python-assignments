````python
"""3.  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance."""
class DiscreteRandomVariable:
    def __init__(self, probabilities, outcomes):
        """
        Initialize the DiscreteRandomVariable with outcomes and their associated probabilities.
        :param probabilities: List of probabilities associated with the outcomes.
        :param outcomes: List of possible outcomes.
        """
        if len(probabilities) != len(outcomes):
            raise ValueError("The length of probabilities and outcomes must be the same.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if not 0 <= sum(probabilities) <= 1:
            raise ValueError("The sum of probabilities must be 1.")
        
        self.probabilities = probabilities
        self.outcomes = outcomes

    def expected_value(self):
        """Calculate the expected value (mean) of the discrete random variable."""
        return sum(p * x for p, x in zip(self.probabilities, self.outcomes))

    def variance(self):
        """Calculate the variance of the discrete random variable."""
        mean = self.expected_value()
        return sum(p * (x - mean) ** 2 for p, x in zip(self.probabilities, self.outcomes))

# Example usage:
probabilities = [1/6] * 6  # Fair six-sided die
outcomes = [1, 2, 3, 4, 5, 6]
random_variable = DiscreteRandomVariable(probabilities, outcomes)

print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

````python
"""5. Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance."""
import numpy as np

def generate_samples(distribution_type, params, size=10000):
    """
    Generate random samples based on the given distribution type and parameters.
    :param distribution_type: Type of distribution ('binomial' or 'poisson').
    :param params: Parameters for the distribution (tuple).
    :param size: Number of random samples to generate.
    :return: A tuple (samples, mean, variance)
    """
    if distribution_type == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif distribution_type == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution type. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    
    return samples, mean, variance

# Example usage:
binomial_samples, binomial_mean, binomial_variance = generate_samples('binomial', (10, 0.5))
poisson_samples, poisson_mean, poisson_variance = generate_samples('poisson', (5,))

print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")
print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
````

````python
"""6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples."""
import numpy as np

def gaussian_samples(mean=0, std_dev=1, size=10000):
    """
    Generate random samples from a Gaussian distribution and compute the mean, variance, and standard deviation.
    :param mean: The mean of the Gaussian distribution.
    :param std_dev: The standard deviation of the Gaussian distribution.
    :param size: Number of random samples to generate.
    :return: A tuple (samples, mean, variance, std_dev)
    """
    samples = np.random.normal(mean, std_dev, size)
    
    mean_val = np.mean(samples)
    variance_val = np.var(samples)
    std_dev_val = np.std(samples)
    
    return samples, mean_val, variance_val, std_dev_val

# Example usage:
samples, mean, variance, std_dev = gaussian_samples(0, 1)

print(f"Gaussian Distribution - Mean: {mean}, Variance: {variance}, Std Dev: {std_dev}")
````

````python
"""8. Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.
"""
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.
    :param x: Value or array of values where PDF is to be calculated.
    :param mean: Mean of the normal distribution.
    :param std_dev: Standard deviation of the normal distribution.
    :return: PDF values for x.
    """
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Example usage
mean = 0
std_dev = 1
x_values = np.linspace(-5, 5, 100)
pdf_values = normal_pdf(x_values, mean, std_dev)

plt.plot(x_values, pdf_values)
plt.title("Normal Distribution PDF")
plt.xlabel("X")
plt.ylabel("PDF")
plt.show()
````

````python
"""9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution."""
def exponential_cdf(x, rate):
    """
    Calculate the Cumulative Distribution Function (CDF) of an exponential distribution.
    :param x: Value or array of values where CDF is to be calculated.
    :param rate: Rate parameter (lambda) of the exponential distribution.
    :return: CDF values for x.
    """
    return 1 - np.exp(-rate * x)

# Example usage
rate = 1  # mean = 1 / rate
x_values = np.linspace(0, 10, 100)
cdf_values = exponential_cdf(x_values, rate)

plt.plot(x_values, cdf_values)
plt.title("Exponential Distribution CDF")
plt.xlabel("X")
plt.ylabel("CDF")
plt.show()
````

````python
"""11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.
To generate the data use the following command:
```python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
  ```
Apply z-test to find which layout is successful."""

import numpy as np
import scipy.stats as stats

# Given data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Conversion rates
p_old = old_layout.mean()  # Proportion of purchases for old layout
p_new = new_layout.mean()  # Proportion of purchases for new layout

# Number of visitors
n_old = len(old_layout)
n_new = len(new_layout)

# Standard error calculation for difference in proportions
se = np.sqrt((p_old * (1 - p_old)) / n_old + (p_new * (1 - p_new)) / n_new)

# Z-statistic
z = (p_new - p_old) / se

# P-value (one-tailed test since we're testing if the new layout is better)
p_value = 1 - stats.norm.cdf(z)

# Output Z-statistic and p-value
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Decision based on significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: New layout leads to higher conversion rate")
else:
    print("Fail to reject null hypothesis: No significant difference")
````

````python
"""15. A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.
Use the following data:
```python
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]"""
import numpy as np
import scipy.stats as stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Calculate the sample means and standard deviations
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)
    std_a = np.std(layout_a_clicks, ddof=1)
    std_b = np.std(layout_b_clicks, ddof=1)
    
    # Number of observations in each group
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)
    
    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n_a - 1) * std_a**2 + (n_b - 1) * std_b**2) / (n_a + n_b - 2))
    
    # Calculate the t-statistic
    t_statistic = (mean_a - mean_b) / (pooled_std * np.sqrt(1/n_a + 1/n_b))
    
    # Degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
    
    return t_statistic, df, p_value

# Data for website layouts
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
t_statistic, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

print(f"T-statistic: {t_statistic}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The layouts have different click-through rates.")
else:
    print("Fail to reject null hypothesis: The layouts have the same click-through rate.")
````

````python
"""16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.
Use the following data of cholestrol level:
```python
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]
```"""
def drug_effect_analysis(existing_drug_levels, new_drug_levels):
    # Calculate the sample means and standard deviations
    mean_existing = np.mean(existing_drug_levels)
    mean_new = np.mean(new_drug_levels)
    std_existing = np.std(existing_drug_levels, ddof=1)
    std_new = np.std(new_drug_levels, ddof=1)
    
    # Number of observations in each group
    n_existing = len(existing_drug_levels)
    n_new = len(new_drug_levels)
    
    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n_existing - 1) * std_existing**2 + (n_new - 1) * std_new**2) / (n_existing + n_new - 2))
    
    # Calculate the t-statistic
    t_statistic = (mean_new - mean_existing) / (pooled_std * np.sqrt(1/n_existing + 1/n_new))
    
    # Degrees of freedom
    df = n_existing + n_new - 2
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
    
    return t_statistic, df, p_value

# Data for cholesterol levels
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform drug effectiveness analysis
t_statistic, df, p_value = drug_effect_analysis(existing_drug_levels, new_drug_levels)

print(f"T-statistic: {t_statistic}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The new drug is more effective than the existing drug.")
else:
    print("Fail to reject null hypothesis: The new drug is not more effective than the existing drug.")
````

````python
"""17. A school district introduces an educational intervention program to improve math scores. Write a Python
function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
determine if the intervention had a significant impact.
Use the following data of test score:
  ```python
  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]"""
def intervention_analysis(pre_intervention_scores, post_intervention_scores):
    # Calculate the differences between pre and post intervention scores
    differences = np.array(post_intervention_scores) - np.array(pre_intervention_scores)
    
    # Calculate the sample mean and standard deviation of differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    
    # Number of observations
    n = len(differences)
    
    # Calculate the t-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))
    
    # Degrees of freedom
    df = n - 1
    
    # Calculate the p-value for a one-tailed test
    p_value = 1 - stats.t.cdf(t_statistic, df)
    
    return t_statistic, df, p_value

# Data for test scores before and after intervention
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform intervention analysis
t_statistic, df, p_value = intervention_analysis(pre_intervention_scores, post_intervention_scores)

print(f"T-statistic: {t_statistic}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The intervention had a significant impact on test scores.")
else:
    print("Fail to reject null hypothesis: The intervention did not significantly impact test scores.")
````

````python
"""18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
significant difference between the average salaries of male and female employees.
Use the below code to generate synthetic data:
```python
# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
  ```
  """
import numpy as np
import scipy.stats as stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def salary_gap_analysis(male_salaries, female_salaries):
    # Calculate means and standard deviations
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)
    
    # Number of observations in each group
    n_male = len(male_salaries)
    n_female = len(female_salaries)
    
    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((n_male - 1) * std_male**2 + (n_female - 1) * std_female**2) / (n_male + n_female - 2))
    
    # Calculate t-statistic
    t_statistic = (mean_male - mean_female) / (pooled_std * np.sqrt(1/n_male + 1/n_female))
    
    # Degrees of freedom
    df = n_male + n_female - 2
    
    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
    
    return t_statistic, df, p_value

# Perform salary gap analysis
t_statistic, df, p_value = salary_gap_analysis(male_salaries, female_salaries)

print(f"T-statistic: {t_statistic}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant gender-based salary gap.")
else:
    print("Fail to reject null hypothesis: There is no significant gender-based salary gap.")
````

````python
"""21. A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.
Use the below code to generate data:
```python
np.random.seed(0)
age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
"""
# Generate data for age groups and voter preferences
np.random.seed(0)

age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a contingency table
contingency_table = np.array([
    [np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '51+') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '51+') & (voter_preferences == 'Candidate B'))]
])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies: \n{expected}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject null hypothesis: There is no significant association between age groups and voter preferences.")
````

````python
"""24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The
company wants to determine if there is a significant difference in customer satisfaction scores among the
three product versions. They conducted a survey and collected customer satisfaction scores for each
version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
difference in customer satisfaction scores.
  Use the following data:
  ```python
  # Sample data: Customer satisfaction scores for each product version
  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
  premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
  deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]"""
# Sample data for customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in customer satisfaction scores between product versions.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in customer satisfaction scores between product versions.")
````
